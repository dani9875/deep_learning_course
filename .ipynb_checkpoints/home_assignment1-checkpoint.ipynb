Exercise 5: 
Randomly selected 32x32 pictures were downloaded from: https://www.iconfinder.com/iconsets/32x32-free-design-icons

Name of files has been collected in one folder and their naming convention are as following: "imgx.png", where 'x' indicates which picture is taken care of, so scanning gets easier.


```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np

np.set_printoptions(threshold=np.nan) #Every element of an array can be seen by setting threshold to nan.

NOP = 5 #Number of pictures to be scanned

ch_R = []
ch_G = []
ch_B = [] #RGB arrays to use in the next exercise

#Iterating the pictures to display them and to print out RGB arrays of those. Please scroll down to the bottom to see all the pictures.
for j in range(NOP):
    img_str = "img" + str(j) + ".png"
    image = plt.imread(img_str) #Numbers in the arrays represent what percentage of each channel is present on a scale of 0 to 255.
    print("\nRGB channels of " + img_str + ", at the bottom, images are depicted:" )
    plt.figure()
    plt.imshow(image)
    i = np.array(image[:,:,0:3]) #Pictures were given in a form of RGBA, thus deleting last column is necessary.
    ch_R.append(image[:,:,0])
    ch_R[j].shape = (1,1024) 
    ch_G.append(image[:,:,1])
    ch_G[j].shape = (1,1024)
    ch_B.append(image[:,:,2])
    ch_B[j].shape = (1,1024)
    print(i)
```

Exercise 6:
First of all 'hand-made' calculations are carried out to get avarages and stds, then Sklearns' StandardScaler helps make sure that the manual calculations are correct.

```python
RGB_avs = np.empty([3,5])
RGB_stds = np.empty([3,5])

scaler = StandardScaler()

for i in range(NOP): #Loop for processing each image
    r = np.array(ch_R[i])
    g = np.array(ch_G[i])
    b = np.array(ch_B[i])
    
    #Calculation of averages and stds
    RGB_avs[0,i] = np.average(r, axis=1)
    RGB_avs[1,i] = np.average(g, axis=1)
    RGB_avs[2,i] = np.average(b, axis=1)
    RGB_stds[0,i] = np.std(r, axis=1)
    RGB_stds[1,i] = np.std(g, axis=1)
    RGB_stds[2,i] = np.std(b, axis=1)
   
    #Result of manual calculations for each picture
    print("\nAvarage:%f, %f, %f and Std:%f, %f, %f of img%d(R,G,B)" % (RGB_avs[0,i],RGB_avs[1,i],RGB_avs[2,i],RGB_stds[0,i],RGB_stds[1,i], RGB_stds[2,i], i) )
    
    #Arranging channels in one array so that standard scaler can be applied to them
    chs = np.array([r,g,b])
    chs = np.transpose(chs)
    chs_scaler =chs.reshape(1024,3)
    scaler.fit(chs_scaler)
    
    #Results with the in-built functions
    print('Average with scaler:[%s]' % ', '.join(map(str, scaler.mean_)))
    print('Std with scaler:[%s]\n' % ', '.join(map(str, np.sqrt(scaler.var_))))
    
    #Standardization with scaler function
    print("Standardized data from left to right ->RGB with scaler\n")
    print(scaler.transform(chs_scaler))
    print("\nStandardization manually:\n")
    

    #In order to standarize, avarage has to be substracted from the data elements which has to be divided by the standard deviation.
    mult = np.ones((1,1024))#It is utilized for manual standardization
    mult *= RGB_avs[0,i]
    r = r - mult
    mult = np.ones((1,1024))
    mult *= RGB_stds[0,i]
    r = r / mult
    r = np.transpose(r)
    
    mult = np.ones((1,1024))
    mult *= RGB_avs[1,i]
    g = g - mult
    mult = np.ones((1,1024))
    mult *= RGB_stds[1,i]
    g = g / mult
    g = np.transpose(g)
    
    mult = np.ones((1,1024))
    mult *= RGB_avs[2,i]
    b = b - mult
    mult = np.ones((1,1024))
    mult *= RGB_stds[2,i]
    b = b / mult
    b = np.transpose(b)
    print(np.array([r,g,b]).reshape(1024,3)) #Printing out the results of manual standardization"
    
```

```python
import scipy.io.wavfile as wf
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(threshold=np.nan)

NFFT = 256 #default value
Fs = 44000 #Sample frequency for audio cd

rate, data = wf.read("out.wav")

d = data[100000:200000,0] #Sample is too long, therefore only one hundred thousands samples are taken into account.
# Data contains two columns since wf.read() returns stereo channels, this is why second column can be omitted.

spectrum, freqs, t, im = plt.specgram(d, NFFT=NFFT,   Fs=Fs, cmap=plt.cm.gist_heat)
plt.ylabel('Frequency')
plt.xlabel('Amplitude')
plt.show() 
```

```python
from bs4 import BeautifulSoup
import urllib.request as urllib2
import re
import pandas as pd
from collections import Counter as cr

url = 'http://www.snopes.com/facebook-ai-developed-own-language/'

content = urllib2.urlopen(url).read()
soup = BeautifulSoup(content,'html.parser')

#For the first try I intended to extract the text from HTML with a simle function call, however soup.get_text()returned some superflous information as well.
#So I decided to consider those texts which belong to a paragraph(<p> ... <\p>). Remark:Title  

phs = soup.find_all('p') #Find all function puts every paragraphs into an array from which only the texts to be extracted.
    
x = soup.title.string #Retrieving the title

web_text_arr = ""
web_text_arr += x #Title is also added to the text

for feat in range(len(phs)):
    web_text_arr = web_text_arr + phs[feat].text

web_text_arr = ''.join(web_text_arr.split()) #Removing all of the whitespace charachters

web_text_arr = re.sub('[^A-Za-z ]+',"", web_text_arr) #Here I use regular expressions to easily remove everything but the letters

#collection library is included to count the letters, but firstly, those have to be lower case because Count function distinguishes upper and lower case
web_text_arr = web_text_arr.lower()
lett = cr(web_text_arr)

#Visualization
df = pd.DataFrame.from_dict(lett, orient='index')
df.plot(kind='bar')

plt.legend(frameon=None)
plt.title("Occurances of the letters(not sorted)")
plt.ylabel('Quantity(piece)')
plt.xlabel('Letters')
plt.show()


```
